Ans:
    
    Difference between a t-test and a z-test:
    Population Variance:

    Z-test: Requires that the population standard deviation (or variance) be known.
    t-test: Does not require the population standard deviation; instead, it uses the sample standard deviation.
    Sample Size:

    Z-test: Typically used when the sample size is large (usually n > 30), as it relies on the Central Limit Theorem for
        large samples.
    t-test: Generally used when the sample size is small (typically n ≤ 30).
    Distribution:

    Z-test: Assumes that the sample means are normally distributed.
    t-test: Assumes that the sample means are normally distributed, but it is less sensitive to violations of normality,
        especially with larger sample sizes.
    Precision:

    Z-test: Generally provides a more precise result when the population standard deviation is known.
    t-test: Provides a more conservative result when the population standard deviation is unknown and estimated from
        the sample.
    Example scenarios:
    Z-test:

    Scenario: Suppose a pharmaceutical company wants to test if a new drug decreases blood pressure. They already have 
        extensive data on the population's blood pressure standard deviation.
    Test: Use a z-test to determine if there's a significant decrease in blood pressure after administering the drug 
        compared to a control group.
        
    t-test:

    Scenario: A teacher wants to determine if a new teaching method improves students' test scores. The teacher selects a
        small sample of students (say, 20 students) and applies the new method to one group while keeping the traditional 
        method for another group.
    Test: Perform a t-test to compare the mean test scores of the two groups and determine if there's a statistically 
        significant difference.


Q2: Differentiate between one-tailed and two-tailed tests.

Ans:
    
    One-Tailed Test:
    Directionality: A one-tailed test is used when you want to determine if a sample mean is significantly greater than or
        less than a known or hypothesized population mean, but not both.

    Hypothesis: The null and alternative hypotheses are set up in such a way that the test focuses on only one direction of
        the distribution.

    Null Hypothesis (H₀): Represents no effect or no difference.
    Alternative Hypothesis (H₁ or Hₐ): Specifies the direction of the effect or difference (either greater than or less than).
    Critical Region: The critical region for the test is located entirely in one tail of the distribution (either the right
                    tail or the left tail).

    Decision: If the test statistic falls into the critical region, you reject the null hypothesis in favor of the 
        alternative hypothesis. Otherwise, you fail to reject the null hypothesis.

    Two-Tailed Test:
    Directionality: A two-tailed test is used when you want to determine if a sample mean is significantly different from 
        a known or hypothesized population mean, regardless of direction (either greater than or less than).

    Hypothesis: The null and alternative hypotheses are set up in a way that the test considers both directions of the
        distribution.

    Null Hypothesis (H₀): Represents no effect or no difference.
    Alternative Hypothesis (H₁ or Hₐ): Specifies that there is a difference, without specifying the direction.
    Critical Region: The critical region for the test is split between both tails of the distribution, capturing extreme
        values in either direction.

    Decision: If the test statistic falls into either of the critical regions (left tail or right tail), you reject the 
        null hypothesis in favor of the alternative hypothesis, suggesting a significant difference. If it falls within
        the non-critical region, you fail to reject the null hypothesis.

Q4: Explain Bayes's theorem with an example.

Ans:
    
    Bayes's Theorem:
    Bayes's Theorem is a fundamental concept in probability theory and statistics that describes the probability of an event
    based on prior knowledge of conditions that might be related to the event. It provides a way to revise existing
    predictions or beliefs in light of new evidence.

    Mathematically, Bayes's Theorem can be represented as:
    P(A∣B)=P(B∣A)×P(A)/P(B)
    
    P(A∣B) is the posterior probability: the probability of hypothesis A given the observed evidence B.
    P(B∣A) is the likelihood: the probability of observing evidence 
    B given that hypothesis A is true.
    P(A) is the prior probability: the initial probability of hypothesis A being true.
    P(B) is the marginal likelihood or evidence: the total probability of observing evidence B.

    Example:
    Let's use a classic example involving a medical diagnostic test to illustrate Bayes's Theorem.

    Suppose there's a rare disease that affects 1% of the population. A diagnostic test for this disease is not perfect:

    If a person has the disease (true positive), the test correctly identifies it 99% of the time (sensitivity).
    If a person does not have the disease (true negative), the test correctly gives a negative result 98% of the time
    (specificity).
    Given this information, let's find out the probability that a person has the disease if the test comes back positive.

    Prior Probability (P(A)): Probability that a randomly selected person from the population has the disease = 0.01
        (or 1%).

    Likelihood (P(B|A)): Probability that the test is positive given that the person has the disease = 0.99 
        (or 99% sensitivity).

    Marginal Likelihood (P(B)): Overall probability of testing positive, which is a combination of true positives and 
        false positives:P(B)=P(B∣A)×P(A)+P(B∣¬A)×P(¬A)
                        P(B)=0.99×0.01+0.02×0.99=0.01+0.0198=0.0298

    (Note:¬A represents not having the disease, and P(B∣¬A) represents the false positive rate, which is 1 - 
     specificity = 0.02.)

    Posterior Probability (P(A|B)): Probability that a person has the disease given that the test is positive:
    P(A∣B)= P(B)P(B∣A)×P(A)
    P(A∣B)= 0.02980.99×0.01≈0.3322
    So, even if the test comes back positive, the probability that a person actually has the disease is approximately 
    33.22%. This example demonstrates how Bayes's Theorem helps update our beliefs based on new evidence, taking into 
    account both the prior probability and the diagnostic accuracy of the test.










Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

Ans:
    
    Confidence Interval:
    A confidence interval (CI) is a range of values derived from sample data that is likely to contain the true population
    parameter with a specified level of confidence. In other words, it provides a range of plausible values for an unknown 
    population parameter, such as a population mean or proportion, based on sample data.

    The confidence level (often denoted as 1−α) represents the probability that the interval estimate will contain the true
    population parameter. Common confidence levels include 90%, 95%, and 99%.

    Calculation of Confidence Interval:
    The formula to calculate the confidence interval for a population mean (μ) using a sample mean ( ˉx), sample standard
    deviation (s), sample size (n), and a specified confidence level is:
    CI= xˉ±Z×( s/root of n)

    Where:
    ˉx= Sample mean
    s = Sample standard deviation
    n = Sample size
    Z = Z-value corresponding to the desired confidence level. For example:
    For a 95% confidence level, Z = 1.96 (for a standard normal distribution).
    For a 99% confidence level, Z = 2.576 (for a standard normal distribution).
    Example:
    Suppose you want to estimate the average weight loss after following a specific diet for 4 weeks. You collect a 
    random sample of 50 individuals who followed this diet, and their average weight loss is 6 pounds with a standard 
    deviation of 2 pounds.

    Sample Mean (ˉx): 6 pounds
    Sample Standard Deviation (s): 2 pounds
    Sample Size (n): 50
    For a 95% confidence interval, using the Z-value of 1.96:
    CI=6±1.96×( 50/root of 2)
    CI=6±1.96×(0.283)
    CI=6±0.554

    This gives the confidence interval as (5.446, 6.554).

    Interpretation: We are 95% confident that the true average weight loss after following the specific diet for 4 weeks
        lies between 5.446 pounds and 6.554 pounds.
